```python
import ipynb
import ipynb.fs.full.Fonctions_metoo
from ipynb.fs.full.Fonctions_metoo import behavior
import pandas as pd


fr_filename='C:/Users/Robin/Dropbox/Travaux_&_Rapports_Stages/UbicomLab-GATech/MeToo/code/data/#balancetonporc/tweet_fr.json'
df_fr = pd.read_json(fr_filename, orient="columns")


en_filename = 'C:/Users/Robin/Dropbox/Travaux_&_Rapports_Stages/UbicomLab-GATech/MeToo/code/data/#metoo/tweet_en.json'
df_en = pd.read_json(en_filename, orient="columns")
```

```python
# create the data frame with only meto or moi aussi tweet
def metoo(df):
    df_metoo=df[df['hashtag'].str.contains(r'metoo|meto|meeto|moiaussi')]
    return(df_metoo)
def no_metoo(df):
    df_nmt=df[~df['hashtag'].str.contains(r'metoo|meto|meeto|moiaussi')] # not contain
    return(df_nmt)
def balanceTonPorc(df):
    df_btp=df[df['hashtag'].str.contains('balancetonporc')] # not contain
    return(df_btp)
```

```python
df_metoo=metoo(df_fr)
df_btp=balanceTonPorc(df_fr)
df_nmt=no_metoo(df_fr)
```
